STEP: 01 : import all necessary libraries


```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

STEP : 02 : build data or import data

```python
stud_data = {
    "Hours_of_study" : [2,1.5,3,7,5,9,2.5,4.5,8,10],
    "Exam_score" : [50,40,55,70,60,90,53,66,85,95]
}
```

Why Linear regression model and not any other?
- because wee are using data and for that linear model shows growth as it predict on the basis of that graph too .

STEP 03: convert data into dataframe using pandas.

```python
df = pd.DataFrame(stud_data)
df
```

STEP 04: Feature Extraction
- now that we have converted our data intp dataframe , now we will exxtract features from it in some varibales to test and train model.


```python
X = df[['Hours_of_study']]
Y = df[['Exam_score']]
```

STEP 05  : setting our model for training and testing

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2 , random_state = 42)
```

STEP 06 : storing our model as a function in a variable so that we don't have to call model with it's actual name instead we'll use its Alias

```python
model = LinearRegression()
```

STEP 07: fitting our model for training


```python
model.fit( X_train , Y_train )
```

User Inpuut for testing

```python
user_input = float(input("Enter number of Hours you study: "))
predicted_score = model.predict([[user_input]])
print(f"Predicted Score: {predicted_score[0][0]:.2f}")
```
